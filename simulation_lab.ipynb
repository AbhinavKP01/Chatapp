```python
import numpy as np
import matplotlib.pyplot as p
k=2
m=1
tmax=10
v=0
x=1.5
tsize=10/100
b=0.2
X=[]
V=[]
A=[]
T=np.arange(0,10.1,0.1)
for t in T:
  X.append(x)
  V.append(v)
  a=((-2*v*b)-(((k)*x)/m))
  A.append(a)
  v=v+(a*tsize)
  x=x+(v*tsize)
p.subplot(4,1,1)
p.plot(T,X)
p.xlabel("Time (s)")
p.ylabel("displacement (m)")
p.subplot(4,1,2)
p.plot(T,V)
p.xlabel("Time (s)")
p.ylabel("Velocity (m/s)")
p.subplot(4,1,3)
p.plot(T,A)
p.xlabel("Time (s)")
p.ylabel("Acceleration (m/s^2)")
p.subplot(4,1,4)
p.plot(X,V)
p.xlabel("Displacement (m)")
p.ylabel("Velocity(m/s)")
```
